- 반복 실험에 대한 평균 및 표준 편차를 계산한다.
- DCGAN에서 저장된 loss output 파일을 이용해 learning curve를 그린다.

```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

- 평균 및 표준편차 계산(a: accuracy, b: recall)

```python
a= [84.39 , 83.18, 80.76]
b= [0.8, 0.88,  0.7267]
```

```python
print(np.round(np.mean(a), 2),'±',np.round(np.std(a), 2))
print(np.round(np.mean(b), 4) * 100,'±',np.round(np.std(b) * 100, 2))
```

- Loss 파일 import 및 plot 그리기

```python
with open('skincancer_gan_benign_Dloss.pkl', 'rb') as filehandle:
    loss_d = pickle.load(filehandle)
with open('skincancer_gan_benign_Gloss.pkl', 'rb') as filehandle:
    loss_g = pickle.load(filehandle)
```

```python
epochs= range(200)
```

```python
loss_g = [np.mean(i) for i in loss_g] 
loss_d = [np.mean(i) for i in loss_d] 
```

```python
plt.subplots(1, figsize=(7,5))
plt.plot(epochs, loss_d, color="r", label="Discriminator Loss")
plt.plot(epochs, loss_g, color="b", label="Generator Loss")
plt.title("DCGAN Learning Curve", pad=30, size=12)
plt.xlabel("Epoch"), plt.ylabel("Loss"), plt.legend(loc="best")
plt.tight_layout()
plt.show()
```
